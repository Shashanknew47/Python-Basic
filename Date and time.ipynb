### timedelta

```python
#calculation of date and time with timedelta
daydelta = datetime.timedelta(days=450); print(daydelta)
hourdelta = datetime.timedelta(hours=12); print(hourdelta) 
weekdelta = datetime.timedelta(weeks=3); print(weekdelta)

T =  datetime.datetime.now()
print('Today:           ',T)

nday = T + daydelta  # additon of days in today's date.
print('Adding 450 days:   ',nday)

week_delta = T + weekdelta
print('Adding 3 weeks:    ',week_delta)

oday = T - hourdelta  
print('deduction 12 hours:',oday)

print('**********************************************************\n')

"calulation of 2 date also give timedelta"

cday = nday-oday+hourdelta

print(cday)  
print(cday.days)    #getting excat number of days    
print(cday.total_seconds())     #Doing calculation in seconds

print(dir(cday))
help(datetime.timedelta)
```

## Intro to Timezone Aware and Not Timezone Aware

- A naive timezone doen not have timezone associated with it. An aware does.

```python
"""
Coordinated Universal Time (or UTC) is the primary time standard by which the world regulates clocks and time. It is within about 1 second of mean solar time at 0° longitude, and is not adjusted for daylight saving time. It is effectively a successor to Greenwich Mean Time (GMT), we call GMT, Zulu time too.

Coordinated Universal Time (UTC) is the basis for civil time today. This 24-hour time standard is kept using highly precise atomic clocks combined with the Earth's rotation
"""
```

```python
#there are 3 ways to get current time.
import pytz

dt_today = datetime.datetime.today()
print((dt_today),': Using today() provides local timezone time.')

dt_now = datetime.datetime.now()
print((dt_now),': Using now() provides timezone aware time only when we pass pytz in this.')  

dt_utcnow = datetime.datetime.utcnow()
print((dt_utcnow),": It returns a datetime object in utc timezone but it's nottimezone aware  \n\n")
```

```python
"""first time zone aware time:
     - To make any datetime timezone aware we need to pass tzinfo=pytz.UTC into it. 
     - It will will return a UTC time. 
     - As you will notice end you see +00.00 which indicates that this is actual UTC time.
"""
dt = datetime.datetime(2016,7,27,12,30,45,tzinfo=pytz.UTC)
print(repr(dt),' : First timezone aware time.\n')
print(dt)
```

```python
'Using now and utcnow with timezone aware time'
print((datetime.datetime.now(tz=pytz.UTC)))

"now() with aware timezone and now() is more easy way instead of utcnow with replace mehtod which does same thing.)"

'converting timezone awaretime with utcnow method which is unnecessary lenghy'
print(datetime.datetime.utcnow().replace(tzinfo=pytz.UTC)) 
```

## Converting utc timezone into other time zones.

```python
"""printing of specific timezone:
    - First we need to create timeaware UTC time.
    - We need to pass utc aware time into astimezone fuction with pytz.timezone('timezone info') info of the time which we want to convert it.
"""
dt_now = datetime.datetime.now(tz=pytz.UTC)
print((dt_now),'current UTC time')

'Specific Aia/Calcutta time in comarizon with UTC time.'
dt_mn = dt_now.astimezone(pytz.timezone('Asia/Calcutta'))
print((dt_mn),' "+5:30 shows that Asia/Calcutta  timezone is 5.30hrs. is ahead of UTC timezone." ')
```
